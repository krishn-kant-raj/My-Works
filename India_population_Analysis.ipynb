```python
allstates = df['State_UT']
min_pop=[]
max_pop=[]
inc_pop=[]
print('Minimum population and Maximum Polpulation is extracted From 1901 to 2011 population dataset')
print('Increase in population is based on difference between max_pop and min_pop\n')

print('States and UT\t\tMin Pop\t\t\t Max Pop\t\tPop Increases')

def state_min_max_population(allstates):
    i=2
    for states in allstates:
        if i<35 and i>1:
            min = round(df.iloc[i,2:].min(),2)
            max = round(df.iloc[i,2:].max(),2)
            min_pop.append(min)
            max_pop.append(max)
            incpop=round((max-min),2)
            inc_pop.append(incpop)
        if len(states)<8:
            print(states, end="\t\t\t")
        elif len(states)>=16:
            print(states, end="\t")
        else:
            print(states, end="\t\t")
            
        if min<99999 or max<99999:
            print(min, end="\t\t\t")
            print(max, end='\t\t')
        else:
            print(min, end="\t\t")
            print(max, end='\t\t')
        if max<99999:
            print(end='\t')
        print(incpop)
        i+=1
    print()
state_min_max_population(allstates)
```
