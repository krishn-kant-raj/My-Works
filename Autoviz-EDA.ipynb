- it helps to minimize the time for Exploratry Data Analysis on doing the different vizualization.

**Below are the code for vizualization**

You don't have need to know much more about the paramentes. Just change the file path and ceperator, other parameters is default. If you wanna customize it the read the documentation [click here](https://pypi.org/project/autoviz/)

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

```python
filename = "F:/Machine Learning/Ml_Data/Advertising.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
